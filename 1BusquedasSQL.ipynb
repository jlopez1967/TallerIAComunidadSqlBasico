Limitar el número de registros a mostrar, muestra aquellos que empleados cuyo primer nombre se repite mas de 10 veces en la tabla empleados

```python
%%sql
select first_name, count(*) from employees group by first_name having count(*) > 10 
```

Limitar el número de registros a mostrar, muestra aquellos que empleados cuyo primer nombre se repite mas de 10 veces en la tabla empleados, ordenaos por su  frecuencia en forma descendente

```python
%%sql
select first_name, count(*) from employees group by first_name having count(*) > 10 order by count(*) desc
```

```python
# Filtrar Resultados

Where, sirve para identificar parámetros deseados en la búsqueda
operadores que se pueden utilizar
=, <>, >, <, >=, <=
Lógicos AND, OR, NOT

Ejemplo: selecciono aquellos empleados que han ganado menos de 62000.00 y los periodos donde ocurrio 
```

```python
%%sql
select emp_no,salary, birth_date, first_name, last_name, gender, hire_date, from_date, to_date, salary
from hr where salary < 62000.00 LIMIT 0,10
```

Ahora queremos ver quienes ganan menos de 62000.00 y que sean de genero femenino

```python
%%sql
select emp_no,salary, birth_date, first_name, last_name, gender, hire_date, from_date, to_date 
from hr where salary < 62000.00 and gender = 'F' LIMIT 0,10
```

Ahora queremos ver quienes ganan menos de 62000.00 y mayor a 55000.00

```python
%%sql
select emp_no,salary, birth_date, first_name, last_name, gender, hire_date,  from_date, to_date 
from hr where  salary < 62000.00 and salary > 55000 LIMIT 0,10
```

Ahora que sea el rango más sencillo con BETWEEN

```python
%%sql
select emp_no,salary, birth_date, first_name, last_name, gender, hire_date, from_date, to_date 
from hr where salary BETWEEN 55000.00 and 62000.00 LIMIT 0,10
```

Si quiero ver la satisfacción del empleado que pertenecen a algun departamento d004 o d005 (utilizamos el where in)

```python
%%sql

select emp_no,salary, birth_date, first_name, last_name, gender, hire_date,dept_no, from_date, to_date 
from hr where dept_no IN ('d005','d004') LIMIT 0,10
```

Salario promedio por departamento

```python
%%sql
select 
dept_no, avg(salary)
from hr
group by dept_no LIMIT 0,10
```
