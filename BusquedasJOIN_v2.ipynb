```python
%%sql 
SELECT 
p1.country, p1.continent, president, prime_minister
from prime_ministers as p1
LEFT join presidents as p2
on p1.country = p2.country;
```

```python
%%sql 
SELECT 
p1.country, p1.continent, president, prime_minister
from prime_ministers as p1
RIGHT join presidents as p2
on p1.country = p2.country;
```
